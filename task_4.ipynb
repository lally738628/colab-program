```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the  quadratic function
def quadratic(x, a=-1, b=2, c=5):
  return a * x**2 + b * x + c
```

```python
# Define the stages of the waterfall model
stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]

# Equally spaced x positions for stages
stage_x_coords = np.linspace(-2, 2, len(stages))

# Compute y positions for stages using the quadratic function
stage_y_coords = quadratic(stage_x_coords)

# Plot the quadratic curve
curve_x_coords = np.linspace(-2.5, 2.5, 500)
curve_y_coords = quadratic(curve_x_coords)

plt.figure(figsize=(10, 6))
plt.plot(curve_x_coords, curve_y_coords, color="blue", linestyle='-')


# Add markers and connect them with vertical lines to create the waterfall effect
for i in range(len(stages)):
    # Draw a vertical drop to simulate a waterfall
    if i > 0: # Avoid drawing line for the first marker
      plt.plot(
          [stage_x_coords[i-1], stage_x_coords[i]],  # Horizontal segment
          [stage_y_coords[i-1], stage_y_coords[i-1]],
           color="blue",
           linewidth=2
      )
      plt.plot(
          [stage_x_coords[i], stage_x_coords[i]], # Vertical drop
          [stage_y_coords[i-1], stage_y_coords[i]],
           color="blue",
           linewidth=2
      )


    # Place markers and labels
    plt.scatter(stage_x_coords[i], stage_y_coords[i], color="red", zorder=5)
    plt.text(stage_x_coords[i], stage_y_coords[i] + 0.5, stages[i], ha="center", fontsize=12)

plt.xlabel("Stages")
plt.ylabel("Value")
plt.title("Waterfall Model Stages")
plt.grid(True)
plt.show()
```

```python
# Style the plot
plt.title("Waterfall Model Representation Using Quadratic Curve", fontsize=14)
plt.xlabel("Stages", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend()
plt.show()
```
